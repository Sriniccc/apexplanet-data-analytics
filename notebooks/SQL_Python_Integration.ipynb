# Subqueries

A subquery is a query written inside another SQL query. Here, we retrieve all transactions where the sales amount is greater than the average sales value.

```python
query = """
SELECT *
FROM sales
WHERE sales >
(
    SELECT AVG(sales)
    FROM sales
);
"""

pd.read_sql(query, conn)
```

# Common Table Expressions (CTEs)

A Common Table Expression (CTE) uses the WITH clause to create a temporary result set that makes complex queries easier to read and maintain.

```python
query = """
WITH branch_sales AS
(
    SELECT
        branch,
        SUM(sales) AS total_sales
    FROM sales
    GROUP BY branch
)

SELECT *
FROM branch_sales;
"""

pd.read_sql(query, conn)
```

# Window Functions

Window functions perform calculations across a set of rows related to the current row without grouping the results. This section demonstrates ROW_NUMBER(), RANK(), LAG(), and LEAD().

## ROW_NUMBER()

Assigns a unique sequential number to each transaction based on sales in descending order.

```python
query = """
SELECT
    invoice_id,
    sales,
    ROW_NUMBER() OVER (ORDER BY sales DESC) AS row_num
FROM sales;
"""

pd.read_sql(query, conn).head()
```

## RANK()

Ranks transactions according to sales. Transactions with the same sales value receive the same rank.## RANK()

Ranks transactions according to sales. Transactions with the same sales value receive the same rank.

```python
query = """
SELECT
    invoice_id,
    sales,
    RANK() OVER (ORDER BY sales DESC) AS sales_rank
FROM sales;
"""

pd.read_sql(query, conn).head()
```

## LAG()

Returns the previous transaction's sales value based on ascending sales order.

```python
query = """
SELECT
    invoice_id,
    sales,
    LAG(sales) OVER (ORDER BY sales) AS previous_sale
FROM sales;
"""

pd.read_sql(query, conn).head()
```

## LEAD()

Returns the next transaction's sales value based on ascending sales order.

```python
query = """
SELECT
    invoice_id,
    sales,
    LEAD(sales) OVER (ORDER BY sales) AS next_sale
FROM sales;
"""

pd.read_sql(query, conn).head()
```

# SQL Views

A SQL View is a virtual table created from a SQL query. Views simplify frequently used queries and improve reusability.

```python
query = """
CREATE VIEW IF NOT EXISTS sales_summary AS
SELECT
    branch,
    SUM(sales) AS total_sales
FROM sales
GROUP BY branch;
"""

conn.execute(query)

pd.read_sql("SELECT * FROM sales_summary;", conn)
```
