```python
plt.figure(figsize=(8,5))

plt.hist(df["sales"], bins=20)

plt.title("Distribution of Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")

plt.show()
```

## Histogram of Sales

### Observation
- Most sales transactions are between 50 and 400.
- High-value sales above 800 are less frequent.
- The sales distribution is right-skewed, indicating more low and medium sales than high sales.

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=df["sales"])

plt.title("Box Plot of Sales")

plt.show()
```

## Box Plot of Sales

### Observation
- Most sales are concentrated within a specific range.
- A few high-value sales appear as outliers.
- These outliers represent unusually large transactions.

```python
plt.figure(figsize=(10,5))

sns.countplot(data=df, x="product_line")

plt.title("Number of Products Sold by Product Line")
plt.xticks(rotation=45)

plt.show()
```

## Bar Chart of Product Line

### Observation
- The number of sales differs across product lines.
- Some product categories are more popular than others.
- This helps identify high-demand product categories.

```python
plt.figure(figsize=(10,8))

sns.heatmap(df.select_dtypes(include=['number']).corr(),
            annot=True,
            cmap="coolwarm")

plt.title("Correlation Heatmap")

plt.show()
```

## Correlation Heatmap

### Observation
- Sales, gross income, tax, and cogs have a strong positive correlation.
- Quantity has a moderate relationship with sales.
- The heatmap helps identify relationships among numerical variables.

```python
plt.figure(figsize=(8,5))

plt.scatter(df["unit_price"], df["sales"])

plt.title("Unit Price vs Sales")
plt.xlabel("Unit Price")
plt.ylabel("Sales")

plt.show()
```

## Scatter Plot

### Observation
- Higher unit prices generally result in higher sales amounts.
- Sales are influenced by both unit price and quantity purchased.
